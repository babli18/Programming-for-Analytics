# Topics - Visualization : Matplotlib and Seaborn

The primary data visualization library in Python is matplotlib, a project begun in the early 2000s, that was built to mimic the plotting capabilities from Matlab. Matplotlib is enormously capable of plotting most things you can imagine, and it gives its users tremendous power to control every aspect of the plotting surface.


That said, it is not the friendliest library for beginners to grasp. Thankfully, pandas makes visualizing data very easy for us and usually plots what we want with a single call to the plot method. pandas does no plotting on its own. It internally calls matplotlib functions to create the plots.


Seaborn is also a visualization library that wraps matplotlib and does not do any actual plotting itself. Seaborn makes beautiful plots and has many types of plots that are not available from matplotlib or pandas. Seaborn works with tidy (long) data, while pandas works best with aggregated (wide) data. Seaborn also accepts pandas DataFrame objects in its plotting functions.


Although it is possible to create plots without ever running any matplotlib code, from time to time, it will be necessary to use it to tweak finer plot details manually.

### Importing the library

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

### Line Plot - Matplotlib 

```python
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'Sales': [15000, 18000, 22000, 21000, 25000, 27000, 30000, 31000, 28000, 26000, 29000, 32000]
}

df = pd.DataFrame(data)
```

```python
# Plot
plt.figure(figsize=(10,5)) #sets the size of the plot
plt.plot(df['Month'],df['Sales'] ,marker='o', linestyle='-', color='b') # plots line graph with 'o' marker. Change the linestyle to '--'
plt.xlabel('X-axis') # X-axis name
plt.ylabel('Y-axis') # Y-axis name 
plt.title('Yearly Sales Data') # Title of the plot
plt.grid(True) # if we want a grid
plt.show() # is used to display all the figures and plots that have been created.
```

### Line Plot - Seaborn

```python
# Plot
plt.figure(figsize=(10, 6))
sns.lineplot(x='Month', y='Sales', data=df, marker='o', color='blue')
plt.xlabel('Month')
plt.ylabel('Sales ($)')
plt.title('Monthly Sales Over a Year')
plt.grid('grey')
plt.show()
```

### Bar Plot - Matplotlib

```python
data = {
    'Channel': ['Social Media', 'TV', 'Radio', 'Newspaper', 'Online'],
    'Spend': [20000, 35000, 15000, 10000, 25000]
}
adv_df = pd.DataFrame(data)
```

```python
# Plot
plt.bar(adv_df['Channel'], adv_df['Spend'], color='skyblue', edgecolor='black')
plt.xlabel('Categories')
plt.ylabel('Spend$')
plt.title('Bar Plot Example - Advertising Spend By Channel')
plt.show()
```

### Bar Plot - Seaborn

```python
# Plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Channel', y='Spend', data=adv_df, palette='viridis',hue = 'Channel')
plt.xlabel('Advertising Channel')
plt.ylabel('Spend ($)')
plt.title('Advertising Spend by Channel')
plt.show()
```

### Horizontal Bar Plot - Matplotlib

```python
# Plot from the same Spend data
plt.barh(adv_df['Channel'], adv_df['Spend'], color='salmon', edgecolor='black')
plt.xlabel('Spend$')
plt.ylabel('Categories')
plt.title('Horizontal Bar Plot - Advertising Spend By Channel')
plt.show()
```

## Horizantal Bar Plot - Seaborn

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Spend', y='Channel', data=adv_df, palette='viridis')
plt.xlabel('Spend ($)')
plt.ylabel('Advertising Channel')
plt.title('Advertising Spend by Channel')
plt.show()
```

## Stacked Bar Plot  - Matplotlib

```python
# Data
categories = ['A', 'B', 'C', 'D']
values1 = [4, 7, 1, 8]
values2 = [5, 6, 2, 7]

# Plot
fig, ax = plt.subplots()
bars1 = ax.bar(categories, values1, label='Group 1', color='royalblue', edgecolor='black')
bars2 = ax.bar(categories, values2, bottom=values1, label='Group 2', color='seagreen', edgecolor='black')

# Add labels, title, and legend
ax.set_xlabel('Categories')
ax.set_ylabel('Values')
ax.set_title('Simple Stacked Bar Plot')
ax.legend() #adds legend to the graph

plt.show()
```

## Scatter Plot - Matplotlib

```python
data = {
    'Customer': ['Customer 1', 'Customer 2', 'Customer 3', 'Customer 4', 'Customer 5'],
    'Acquisition_Cost': [300, 500, 700, 200, 400],
    'Revenue': [1500, 2500, 3500, 1000, 2000]
}
df = pd.DataFrame(data)
```

```python
# Plot
plt.figure(figsize=(10,6))
plt.scatter(df['Acquisition_Cost'],df['Revenue'], color='orange') #creates the scatter plot
plt.xlabel('Customer Acquisition Cost ($)')
plt.ylabel('Revenue ($)')
plt.title('Scatter Plot Example - Customer Acquisition Cost vs. Revenue')
plt.grid('Grey')
plt.show()
```

## Scatter plot - Seaborn

```python
# Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Acquisition_Cost', y='Revenue', data=df, s=100, color='g', marker='o')
plt.xlabel('Customer Acquisition Cost ($)')
plt.ylabel('Revenue ($)')
plt.title('Customer Acquisition Cost vs. Revenue')
plt.grid(True)
plt.show()
```

```python
# Slightly more complicated scatter plot

import numpy as np
# Data
np.random.seed(42)
x = np.random.randn(100) # x and y are arrays of 100 random numbers each, 
y = np.random.randn(100) # representing the coordinates of the scatter plot.
colors = np.random.rand(100) # colors is an array of 100 random numbers, representing the color of each point.


# Plot
plt.figure(figsize=(10, 6))
#c=colors assigns colors to each point.
#alpha=0.5 sets the transparency of the points.
#cmap='viridis' sets the colormap to 'viridis'.
scatter = plt.scatter(x, y,c=colors, alpha=0.5, cmap='viridis') 
plt.colorbar(scatter, label='Color Scale')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Complex Scatter Plot Example')
plt.grid(True)
plt.show()
```

## Histogram - Matplotlib

```python
# Data
np.random.seed(10)
histdata = np.random.normal(0, 1, 1000)  # 1000 values from a normal distribution

# Plot
plt.figure(figsize=(10, 6))
plt.hist(histdata, bins=30, color='purple', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Normally Distributed Data')

plt.show()
```

## Histogram - Seaborn


```python
# Plot
plt.figure(figsize=(10, 6))
sns.histplot(histdata, bins=30, kde=True, color='purple')
plt.xlabel('Sales ($)')
plt.ylabel('Frequency')
plt.title('Distribution of Sales')
plt.show()
```

## Box Plot - Matplotlib


```python
data = {
    'Region': ['North', 'South', 'East', 'West'] * 5,
    'Sales': [25000, 27000, 30000, 32000, 15000, 18000, 22000, 24000, 20000, 22000, 25000, 27000, 10000, 15000, 18000, 21000, 28000, 31000, 35000, 37000]
}

regiondf = pd.DataFrame(data)
```

```python
# Prepare data for Matplotlib
regions = regiondf['Region'].unique()
data_to_plot = [regiondf[regiondf['Region'] == region]['Sales'] for region in regions]

# Plot
plt.figure(figsize=(10, 6))
plt.boxplot(data_to_plot, labels=regions, patch_artist=True)

# Customization
plt.xlabel('Region')
plt.ylabel('Sales ($)')
plt.title('Sales Distribution by Region')
plt.grid(True, linestyle='--', alpha=0.7)

# Show plot
plt.show()
```

```python
# Plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Region', y='Sales', data=regiondf, palette='muted')
plt.xlabel('Region')
plt.ylabel('Sales ($)')
plt.title('Sales Distribution by Region')
plt.show()
```
